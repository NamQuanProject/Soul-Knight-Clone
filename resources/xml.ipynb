```python
import xml.etree.ElementTree as ET

def extract_bndbox_from_xml(file_path):
    try:
        # 開啟XML檔案並讀取資料
        with open(file_path, 'r') as file:
            xml_data = file.read()
            
            # 解析XML
            root = ET.fromstring(xml_data)

            # 創建字典，用於將不同的名稱映射到對應的列表
            bndbox_dict = {}

            # 找到所有的<object>元素
            objects = root.findall('object')
            for obj in objects:
                # 找到<name>元素
                name = obj.find('name').text

                # 找到<bndbox>元素
                bndbox = obj.find('bndbox')
                if bndbox is not None:
                    # 提取xmin、ymin、xmax、ymax資料
                    xmin = int(bndbox.find('xmin').text)
                    ymin = int(bndbox.find('ymin').text)
                    xmax = int(bndbox.find('xmax').text)
                    ymax = int(bndbox.find('ymax').text)

                    # 將提取的資料存入對應的列表
                    if name not in bndbox_dict:
                        bndbox_dict[name] = []
                    bndbox_dict[name].append((xmin, ymin, xmax, ymax))

            return bndbox_dict
        
    except FileNotFoundError:
        print("找不到指定的檔案:", file_path)
        return {}
```

```python
# 指定XML檔案路徑
xml_file_path = "./1-1.xml"

# 提取資料
bndbox_dict = extract_bndbox_from_xml(xml_file_path)
for tag in bndbox_dict.items():
    if (tag[0] == "wall"):
        print(f"-----------------------------{tag[0]}-----------------------------------")
        for box in tag[1]:
            print(f"AddWall(Point({box[0]}, {box[1]}), Point({box[2]}, {box[3]}));")
    else:
        print(f"-----------------------------{tag[0]}-----------------------------------")
        for box in tag[1]:
            print(f"Point({box[0]}, {box[1]}), Point({box[2]}, {box[3]})")


```
